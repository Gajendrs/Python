```python
#wap find factorail of any number using for 
# 5! --> 5*4*3*2*1=
N = input("enter any number ")
N =int(N)
if(N<0):
    print("for -ve numberws factorial is not possible ")
elif(N == 0 or N ==1):
    print("fact of ",N ,"is 1")
else:
    fact =1
    for e in range(2,N+1):
        fact = fact*e
    print("fact of ",N,"! = ",fact)
```
